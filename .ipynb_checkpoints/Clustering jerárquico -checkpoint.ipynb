# Clustering jerárquico y dendograma
 

1. dataset array(N*M) de puntos a clusterizar
2. N: número de datos
3. M: número de rasgos
4. Z: array de enlace del cluster con información de las uniones
5. K: número de clusters

```python
import matplotlib.pyplot as plt 
from scipy.cluster.hierarchy import dendrogram, linkage
import numpy as np
```

```python
np.random.seed(4711)
a = np.random.multivariate_normal([10,0], [[3,1], [1,4]], size=[100,])
b = np.random.multivariate_normal([0,20], [[3,1], [1,4]], size=[50,])
X = np.concatenate((a,b))
print(X.shape)
plt.scatter(X[:,0], X[:,1])
```

```python
Z = linkage(X, "ward")
Z
```

```python
from scipy.cluster.hierarchy import cophenet
from scipy.spatial.distance import pdist 
```

```python
c, coph_dist = cophenet(Z,pdist(X))
c
```

### Representación gráfica de dendograma

```python
plt.figure(figsize=(30,20))
plt.title("Dedograma de clustering jerarquico")
plt.xlabel("Indices de la muestra")
plt.ylabel("Distancia")
dendrogram(Z, leaf_rotation = 90, leaf_font_size=8.0, color_threshold=0.1*180.27)
plt.show()
```

### Truncar el drendograma

```python
plt.figure(figsize=(30,20))
plt.title("Dedograma de clustering jerarquico truncado")
plt.xlabel("Indices de la muestra")
plt.ylabel("Distancia")
dendrogram(Z, leaf_rotation = 90, leaf_font_size=12.0, color_threshold=0.1*180.27, truncate_mode="lastp", p=12
          , show_leaf_counts=False, show_contracted=True)
plt.show()
```

### Drendrograma personalizado

```python
def dendrogram_per(*args, **kwargs): 
    max_d = kwargs.pop("max_d", None)
    if max_d and 'color_threshold' not in kwargs:
        kwargs['color_threshold']=max_d
    annotate_above = kwargs.pop('annotate_above', 0)
    ddata = dendrogram(*args, **kwargs)
    if not kwargs.get('no_plot', False): 
        plt.title("Clustering jerarquico con Drendograma truncado")
        plt.xlabel("Indice del dataset (tamaño del cluster)")
        plt.ylabel("Distancia")
        for i,d,c in zip(ddata['icoord'],ddata['dcoord'], ddata['color_list']):
            x= 0.5*sum(i[1:3])
            y= d[1]
            if y > annotate_above:
                plt.plot(x,y,'o', c=c)
                plt.annotate('%.3g'%y, (x,y), xytext=(0,-5),
                           textcoords="offset points", va="top", ha="center")
    if max_d:
        plt.axhline(y=max_d, c='k')
        
    return ddata
```

```python
dendrogram_per(Z, truncate_mode = "lastp", p=12, leaf_rotation =90, leaf_font_size=12.0, show_contracted=True, annotate_above=10, max_d=20)
plt.show()
```

### Corte automatico del dendrograma
inconsistecy_i = (h_i-avg(h_j)) /std(h_j)

```python
from scipy.cluster.hierarchy import inconsistent
```

```python
depth = 3
inconse = inconsistent(Z, depth)
inconse[-10:]
```

### Método del codo

```python
last = Z[-10:,2] #10 ultimos, segunda columna
print(last)
last_rev = last[::-1] # voltearlos
idx = np.arange(1, len(last)+1)
print(last_rev)
plt.plot(idx, last_rev)

acc = np.diff(last, 2) # restar cada 1 con el siguiente
acc_rev = acc[::-1] 
plt.plot(idx[:-2]+1,acc_rev) #en lugar de pintar del 1 al 8, pintamos del 2 al 9
plt.show()
k = acc_rev.argmax() +2 
print("El numero optimo de clusters es %s",str(k))
```

```python
c = np.random.multivariate_normal([40,40], [[20,1], [1,30]], size = [200,])
d = np.random.multivariate_normal([80,80], [[30,1], [1,30]], size = [200,])
e = np.random.multivariate_normal([0,100], [[100,1], [1,100]], size = [200,])
X2 = np.concatenate((X,c,d,e),)
plt.scatter(X2[:,0],X2[:,1])
plt.show()
```

```python
Z2 = linkage(X2, "ward")
plt.figure(figsize=(10,10))
dendrogram_per(Z2, truncate_mode = "lastp", p=30, leaf_rotation=90, leaf_font_size=10.0, show_contracted = True, 
             annotate_above=40, max_d = 170)
plt.show()
```

```python
last = Z2[-10:,2] #10 ultimos, segunda columna
print(last)
last_rev = last[::-1] # voltearlos
idx = np.arange(1, len(last)+1)
print(last_rev)
plt.plot(idx, last_rev)

acc = np.diff(last, 2) # restar cada 1 con el siguiente
acc_rev = acc[::-1] 
plt.plot(idx[:-2]+1,acc_rev) #en lugar de pintar del 1 al 8, pintamos del 2 al 9
plt.show()
k = acc_rev.argmax() +2 
print("El numero optimo de clusters es %s",str(k))
```

### Recuperar los clusters y sus elementos

```python
from scipy.cluster.hierarchy import fcluster
```

```python
max_d = 20
clusters = fcluster(Z, max_d, criterion = "distance")
clusters
```

```python
k = 3
clusters = fcluster(Z, k, criterion="maxclust")
clusters
```
