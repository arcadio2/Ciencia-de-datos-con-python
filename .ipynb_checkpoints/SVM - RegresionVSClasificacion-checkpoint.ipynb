# Clasificacion de flores de iris

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
from sklearn import svm, datasets
%matplotlib inline
```

```python
iris = datasets.load_iris()
print(iris)
```

```python
X = iris.data[:, :2]
Y = iris.target
```

```python
x_min, x_max = X[:,0].min()-1, X[:,0].max()+1
y_min, y_max = X[:,1].min()-1, X[:,1].max()+1
h = (x_max - x_min)/100

xx,yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

X_plot = np.c_[xx.ravel(), yy.ravel()]
```

```python
X_plot
```

```python
def rep(X,Y, kernel):
    C = 1.0
    svc = svm.SVC(kernel = kernel, C=C, decision_function_shape = "ovr").fit(X,Y)
    Y_pred = svc.predict(X_plot)
    Y_pred = Y_pred.reshape(xx.shape)
    plt.figure(figsize=(16,9))
    plt.contourf(xx,yy,Y_pred, cmap = plt.cm.tab10, alpha = 0.3)
    plt.scatter(X[:,0], X[:,1], c=Y,cmap = plt.cm.tab10)
    plt.xlabel("Longitud de los petalos")
    plt.ylabel("Anchura de los petalos")
    plt.xlim(x_min,x_max)
    plt.title("SVC para las flores de Iris con Kernel: "+kernel)
```

```python
rep(X=X,Y=Y,kernel="linear")
```

```python
rep(X=X,Y=Y,kernel="rbf")
```

```python
rep(X=X,Y=Y,kernel="sigmoid")
```

```python
rep(X=X,Y=Y,kernel="poly")
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import classification_report
from sklearn.utils import shuffle
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y ,test_size=0.5, random_state=0)
```

```python
parameters = [
    {
        'kernel':['rbf'],
        'gamma' : [1e-4, 1e-3, 1e-2,0.1,0.2,0.5],
        'C': [1,10,100,1000]
    },
    {
        'kernel':['rbf'],
        'C': [1,10,100,1000]
    }
]
```

```python
clf = GridSearchCV(svm.SVC(decision_function_shape = "ovr"), param_grid=parameters, cv=5)
clf.fit(X,Y)
```

```python
clf.best_params_
```

```python
means = clf.cv_results_['mean_test_score']
stds = clf.cv_results_['std_test_score']
params = clf.cv_results_['params']

for m,s,p in zip(means, stds, params): 
    print("%0.3f (+/-%0.3f) para %r"%(m,s*2,p))
```

```python
y_pred = clf.predict(X_test)
print(classification_report(Y_test, y_pred, target_names = ["setosa", "versicolor", "virginica"]))
```

# Resumen final de la clasificación de iris

```python
def svm_iris(C=1.0, gamma = 0.01, kernel = "rbf", ):
    import pandas as pd
    import numpy as np
    import matplotlib.pyplot as plt 
    from sklearn import svm, datasets
    
    iris = datasets.load_iris()
    
    X = iris.data[:, :2]
    Y = iris.target
    
    x_min, x_max = X[:,0].min()-1, X[:,0].max()+1
    y_min, y_max = X[:,1].min()-1, X[:,1].max()+1
    h = (x_max - x_min)/100

    xx,yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

    svc = svm.SVC(kernel = kernel, C=C, gamma = gamma, decision_function_shape = "ovr").fit(X,Y)
    Y_pred = svc.predict(X_plot)
    Y_pred = Y_pred.reshape(xx.shape)
    plt.figure(figsize=(16,9))
    plt.contourf(xx,yy,Y_pred, cmap = plt.cm.tab10, alpha = 0.3)
    plt.scatter(X[:,0], X[:,1], c=Y,cmap = plt.cm.tab10)
    plt.xlabel("Longitud de los petalos")
    plt.ylabel("Anchura de los petalos")
    plt.xlim(x_min,x_max)
    plt.title("SVC para las flores de Iris con Kernel: "+kernel)
```
