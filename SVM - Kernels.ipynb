# Identificar fronteras no lineales

```python
from sklearn.datasets import make_circles, make_blobs
import matplotlib.pyplot as plt 
from sklearn.svm import SVC
import numpy as np
```

```python
X, Y =make_circles(100, factor=.1, noise=.1)
```

```python
def plt_svc(model, ax=None, plot_support=True): 
    """Plot de la función de decisión para una clasificación en 2D con SVC"""
    if ax == None: 
        ax = plt.gca()
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()
    
    #Generamos los puntos para evaluar el modelo
    xx = np.linspace(xlim[0], xlim[1], 30)
    yy = np.linspace(ylim[0], ylim[1], 30)
    Y, X = np.meshgrid(yy,xx)
    
    xy = np.vstack([X.ravel(), Y.ravel()]).T
    P = model.decision_function(xy).reshape(X.shape)
    
    print(model.support_vectors_)
    ##Representamos las fronteras y margenes del SVC 
    ax.contour(X,Y,P, colors="k", levels = [-1,0,1], linestyles=["--", "-", "--"])
    
    if plot_support:
        ax.scatter(model.support_vectors_[:,0], model.support_vectors_[:,1], s=300, linewidth=1,facecolors="purple")
        
    ax.set_xlim(xlim)
    ax.set_ylim(ylim)
```

```python
def plt_svm(N=10, ax=None): 
    X, Y = make_blobs(n_samples=200, centers=2, random_state=0,cluster_std=0.6)
    X = X[:N]
    Y = Y[:N]
    model = SVC(kernel="linear", C = 1E10)
    model.fit(X,Y)
    
    ax = ax or plt.gca()
    ax.scatter(X[:,0], X[:,1], c=Y, s=50, cmap="autumn")
    ax.set_xlim(-1,4)
    ax.set_ylim(-1,6)
    plt_svc(model, ax)
```

```python
plt.scatter(X[:,0], X[:,1], c=Y)
plt_svc(SVC(kernel="linear").fit(X,Y), plot_support=False)
```

```python
from mpl_toolkits import mplot3d
```

```python
def plot_3D(elev=30, azim=30, X=X,Y=Y,r=r):
    ax = plt.subplot(projection="3d")
    ax.scatter3D(X[:,0], X[:,1], r, c=Y, s=50)
    ax.view_init(elev=elev, azim=azim)
    
    ax.set_xlabel("x[0]")
    ax.set_ylabel("y[1]")
    ax.set_zlabel("r")
```

```python
from ipywidgets import interact, fixed
```

```python
interact(plot_3D, elev=[-90,-60,-30,0,30,60,90], 
         azim=[-180,-150,-120,-90,-60,-30,0,30,60,90, 120,150,180], X = fixed(X), Y = fixed(Y), r = fixed(r))
```

## Ajustar los parametros de SVM

```python
X, Y = make_blobs(n_samples = 100, centers = 2, random_state=0, cluster_std = 1.2)
```

```python
plt.scatter(X[:,0], X[:,1], s=50, c=Y, cmap="autumn")
model = SVC(kernel="linear", C=1E6)
model.fit(X,Y)
plt_svc(model)
```

```python
X, Y = make_blobs(n_samples = 100, centers = 2, random_state=0, cluster_std = 0.8)
plt.scatter(X[:,0], X[:,1], s=50, c=Y, cmap="autumn")
```

```python
fig, ax = plt.subplots(1,2, figsize=(16,6))
fig.subplots_adjust(left = 0.05, right = 0.95, wspace=0.1)

for ax_i, C in zip(ax, [10.0, 0.1]):
    model = SVC(kernel="linear", C=C)
    model.fit(X,Y)
    ax_i.scatter(X[:,0], X[:,1], c=Y, s=50, cmap="autumn")
    plt_svc(model, ax_i)
    ax_i.set_title("C = {0:.1f}".format(C), size=15)
```
