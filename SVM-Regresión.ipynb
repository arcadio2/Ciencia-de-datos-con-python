```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVR
```

```python
X = np.sort(5*np.random.rand(200, 1), axis=0)
Y = np.sin(X).ravel()
#print(list(zip(X.tolist(), Y.tolist())))
Y[::5] += 3*(0.5- np.random.rand(40))  #de 5 en 5 paa los 200
```

```python
plt.scatter(X,Y, color="darkorange", label="data")
```

```python
C = 1e3
svr_lin = SVR(kernel="linear", C=C)
svr_rbf = SVR(kernel="rbf", C=C, gamma=0.01)
svr_pol = SVR(kernel="poly", C=C, degree=3)
```

```python
y_lin = svr_lin.fit(X,Y).predict(X)
y_rbf = svr_rbf.fit(X,Y).predict(X)
y_pol = svr_pol.fit(X,Y).predict(X)
```

```python
lw = 2
plt.figure(figsize=(16,9))
plt.scatter(X,Y, color="darkorange", label="data")
plt.plot(X,y_lin, color="navy", lw=lw, label="SVR linear")
plt.plot(X,y_rbf, color="c", lw=lw, label="SVR radial")
plt.plot(X,y_pol, color="cornflowerblue", lw=lw, label="SVR Polinomico")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("SUPPORT VECTOR REGRESION")
plt.legend()
plt.show()
```
